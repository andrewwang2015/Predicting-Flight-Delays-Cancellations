```python
test_dt = '2016-01-02 03:35:00'

#test_dt_a = datetime.strptime(test_dt, format='%Y-%m-%d %H:%M')
def getClosestWeather(date_str, airport_code, hour_radius):
    test_dt = datetime.datetime(*(time.strptime(date_str,'%Y-%m-%d %H:%M:%S')[0:6]))

    start_dt = test_dt - datetime.timedelta(hours=hour_radius)
    end_dt = test_dt + datetime.timedelta(hours=hour_radius)

    start_dt_str = start_dt.strftime('%Y-%m-%d %H:%M')
    end_dt_str = end_dt.strftime('%Y-%m-%d %H:%M')
    
    airport_df_dict = weather_df_tuples[airport_code]
    #print (airport_df_dict)
    
    if start_dt_str[:10] == end_dt_str[:10]:
        if start_dt_str[:4] == '2014':
            start_dt_str = '2015-01-01'
        if start_dt_str[:10] == '2016-01-02':
            start_dt_str = '2016-01-01'
        if end_dt_str[:10] == '2016-01-02':
            end_dt_str = '2016-01-01 23:59:00'
        airport_df = airport_df_dict[start_dt_str[:10]]
    else:
        if start_dt_str[:4] == '2014':
            start_dt_str = '2015-01-01'
        if start_dt_str[:10] == '2016-01-02':
            start_dt_str = '2016-01-01'
        if end_dt_str[:10] == '2016-01-02':
            end_dt_str = '2016-01-01 23:59:00'
        l_of_d = [airport_df_dict[start_dt_str[:10]], airport_df_dict[end_dt_str[:10]]]
        airport_df = pd.concat(l_of_d)
        
    mask = (airport_df['valid'] > start_dt_str) & (airport_df['valid'] <= end_dt_str)

    close_dts = airport_df.loc[mask]

    str_dates = close_dts['valid'].tolist()
    dates = []
    for date in str_dates:
        dates.append(datetime.datetime(*(time.strptime(date,'%Y-%m-%d %H:%M')[0:6])))
        
    if dates == []:
        print (date_str, ' ', airport_code)
        return getClosestWeather(date_str, airport_code, hour_radius + 1)

    final_dt = min(dates, key=lambda d: abs(d - test_dt))
    return final_dt.strftime('%Y-%m-%d %H:%M')

print(getClosestWeather(test_dt, 'ANC', 1))
```
