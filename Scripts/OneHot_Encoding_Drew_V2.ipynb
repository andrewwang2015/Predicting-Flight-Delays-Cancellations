```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
import time
import glob
%matplotlib inline
from sklearn import preprocessing
import tables
import h5py
from sqlalchemy import create_engine
```

```python
flights = pd.read_csv('BigData_Pruned_drop_Manu_V1.csv')
flights.head()
```

```python
list(flights)
```

```python
flights["MONTH"] = flights["MONTH"].astype(int)
```

```python
flights["DOW"] = flights["DOW"].astype(int)
```

```python
flights["ARRIVAL_DELAY"] = flights["ARRIVAL_DELAY"].astype(int)
```

```python
flights["DIVERTED"] = flights["DIVERTED"].astype(int)
```

```python
flights["CANCELLED"] = flights["CANCELLED"].astype(int)
```

```python
flights["dept_windspd"] = flights["dept_windspd"].astype(int)
flights["arr_windspd"] = flights["arr_windspd"].astype(int)
```

```python
flights["dept_precip"] = flights["dept_precip"].astype(int)
flights["arr_precip"] = flights["arr_precip"].astype(int)
```

```python
flights["dept_vsby"] = flights["dept_vsby"].astype(int)
flights["arr_vsby"] = flights["arr_vsby"].astype(int)
```

```python
flights.to_csv("BigData_Pruned_drop_Manu_V2.csv", index = False)
```

```python
orig_airports = flights['ORIGIN_AIRPORT'].tolist()
dest_airports = flights['DESTINATION_AIRPORT'].tolist()
```

```python
le_airports = preprocessing.LabelEncoder()
le_airports.fit(orig_airports)
#print(le_airports.classes_)
enc_orig_airports = le_airports.transform(orig_airports)
enc_dest_airports = le_airports.transform(dest_airports) 

enc_dest_airports
```

```python
flights['ORIGIN_AIRPORT'] = enc_orig_airports
flights['DESTINATION_AIRPORT'] = enc_dest_airports
flights.head()
```

```python
manufacts = flights['Manu'].tolist()

le_manu = preprocessing.LabelEncoder()
le_manu.fit(manufacts)
#print(le_airports.classes_)
enc_manufacts = le_manu.transform(manufacts)

flights['Manu'] = enc_manufacts
flights.head()
```

```python
models = flights['Model'].tolist()

le_model = preprocessing.LabelEncoder()
le_model.fit(models)
#print(le_airports.classes_)
enc_models = le_model.transform(models)

flights['Model'] = enc_models
flights.head()
```

```python
arr_weather = flights['arr_weather'].tolist()
dept_weather = flights['dept_weather'].tolist()

le_arr_weather = preprocessing.LabelEncoder()
le_arr_weather.fit(arr_weather)
#print(le_airports.classes_)
enc_arr_weather = le_arr_weather.transform(arr_weather)

le_dept_weather = preprocessing.LabelEncoder()
le_dept_weather.fit(dept_weather)
#print(le_airports.classes_)
enc_dept_weather = le_dept_weather.transform(dept_weather)

flights['arr_weather'] = enc_arr_weather
flights['dept_weather'] = enc_dept_weather
flights.head()
```

```python
arr_sky = flights['arr_sky'].tolist()
dept_sky = flights['dept_sky'].tolist()

le_sky = preprocessing.LabelEncoder()
le_sky.fit(arr_sky)
#print(le_airports.classes_)
enc_arr_sky = le_sky.transform(arr_sky)
enc_dept_sky = le_sky.transform(dept_sky)

flights['dept_sky'] = enc_dept_sky
flights['arr_sky'] = enc_arr_sky
flights.head()
```

```python
airlines = flights['AIRLINE'].tolist()
le_airline = preprocessing.LabelEncoder()
le_airline.fit(airlines)
#print(le_airports.classes_)
enc_airlines = le_airline.transform(airlines)

flights['AIRLINE'] = enc_airlines
flights.head()
```

```python
flights.loc[flights.arr_gust == 'M', 'arr_gust'] = 0
flights.loc[flights.dept_gust == 'M', 'dept_gust'] = 0
flights.head()
```
