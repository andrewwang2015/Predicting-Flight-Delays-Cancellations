```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
from datetime import date
%matplotlib inline
```

```python
flights = pd.read_csv('biggestData.csv', index_col = False)
```

```python
flights.head()
```

```python
del flights["Unnamed: 0"]
```

```python
monthMap = {
    '01- January':1,
    '02- February':2,
    '03- March':3,
    '04- April':4,
    '05- May':5,
    '06- June':6,
    '07- July':7,
    '08- August':8,
    '09- September':9,
    '10- October':10,
    '11- November':11,
    '12- December':12
    
}
flights['MONTH_NAME'] = flights['MONTH_NAME'].apply(lambda d: monthMap[d])
```

```python
dayOfWeekMap = {
    'Monday':1,
    'Tuesday':2,
    'Wednesday':3,
    'Thursday':4,
    'Friday':5,
    'Saturday':6,
    'Sunday':7
}
flights['DOW_NAME'] = flights['DOW_NAME'].apply(lambda d: dayOfWeekMap[d])
```

```python
flights.head()
```

```python
flights['DEPARTURE_SCHEDULED'] = pd.to_datetime(flights['DEPARTURE_SCHEDULED'])
flights['ARRIVAL_SCHEDULED'] = pd.to_datetime(flights['ARRIVAL_SCHEDULED'])
```

```python
daysSinceBeginning = []
d0 = date(2015,1,1)
delta = flights['DEPARTURE_SCHEDULED'] - d0
for i in delta:
    daysSinceBeginning.append(i.days)
```

```python
departureHours = []
for i in delta:
    departureHours.append(round(i.seconds/(3600),1))
```

```python
arrivalHours = []
d0 = date(2015,1,1)
delta2 = flights['ARRIVAL_SCHEDULED'] - d0
for i in delta2:
    arrivalHours.append(round(i.seconds/(3600),1))
    
```

```python
flights["DAYS"] = daysSinceBeginning
```

```python
flights["DEPARTURE_HOUR"] = departureHours
flights["ARRIVAL_HOUR"] = arrivalHours
```

```python
flights.head()
```

```python
del flights["DEPARTURE_SCHEDULED"]
del flights["ARRIVAL_SCHEDULED"]
```
